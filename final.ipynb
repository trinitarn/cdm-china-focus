```python
import pandas as pd
filtered_df = df[df["List of host countries (ISO 2)"] == "CN"]

print(filtered_df)
```

```python
grouped_sectors = filtered_df.groupby("Project type (UNEP DTU)").size()

print(grouped_sectors)
```

```python
import pandas as pd
grouped_emissions = filtered_df.groupby("Project type (UNEP DTU)")["Amount Of Reductions (PDD ex-ante) per year in CP1 (PoA based on the sum of curr"].sum()

top_10_emissions = grouped_emissions.nlargest(10)
labels = top_10_emissions.index.to_numpy()
values = top_10_emissions.to_numpy()

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Top 10 Sectors with Highest Emissions Reduced")
plt.axis("equal")
plt.show()
```

```python
## 1c the reducement of emission per year
```

```python
year_columns = ["Amount Of Reductions (PDD ex-ante) per year in CP1 (PoA based on the sum of curr", "Amount Of Reductions (PDD ex-ante) per year in CP2", "Amount Of Reductions (PDD ex-ante) per year in CP3"]

# Group data by sector and calculate the sum of emissions for each year
emission_yearly = filtered_df.groupby("Project type (UNEP DTU)")[year_columns].sum()

# Create separate line charts for each sector
for sector, emissions in emission_yearly.iterrows():
    plt.figure(figsize=(10, 6))
    plt.plot(emissions, marker="o")
    plt.title(f"Emission Reduction for {sector}")
    plt.xlabel("Year")
    plt.ylabel("Emission Reduction")
    plt.xticks(range(1, len(year_columns) + 1), year_columns)
    plt.grid(True)
    plt.show()
```

```python
year_columns = ["Amount Of Reductions (PDD ex-ante) per year in CP1 (PoA based on the sum of curr", "Amount Of Reductions (PDD ex-ante) per year in CP2", "Amount Of Reductions (PDD ex-ante) per year in CP3"]

# Group data by sector and calculate the sum of emissions for each year
emission_yearly = filtered_df.groupby("Project type (UNEP DTU)")[year_columns].sum()
print(emission_yearly)
```

```python
# Create a line chart with separate lines for each sector
plt.figure(figsize=(10, 6))
emission_yearly.plot(kind="line", marker="o")
plt.title("Emission Reduction by Sector Over Time")
plt.xlabel("Year")
plt.ylabel("Emission Reduction")
plt.legend(title="Sector")
plt.grid(True)
plt.xticks(range(1, len(year_columns) + 1), year_columns)
plt.show()
```

```python
df=pd.read_csv("datasett.csv")
import pandas as pd

df['PoA/Lastest CPA request'] = pd.to_datetime(df['PA:Lastest registration request/PoA:Lastest CPA request'], format='%Y-%m-%d')
df['Registered or rejected'] = pd.to_datetime(df['Registered or rejected'], format='%Y-%m-%d')
df['Days Difference'] = (df['Registered or rejected'] - df['PoA/Lastest CPA request']).dt.days

print(df)
```

```python
df['Country'] = df['List of host countries (ISO 2)'].str.split(';')
df = df.explode('Country')

grouped_data = df.groupby('Country')['Days Difference'].mean().sort_values(ascending=False)
grouped_data=grouped_data.dropna()
top_10_countries = grouped_data.head(10)

print(top_10_countries)
```

```python
grouped_data = grouped_data[grouped_data > 0]
last_10_countries = grouped_data.tail(10)
print(last_10_countries)
```
